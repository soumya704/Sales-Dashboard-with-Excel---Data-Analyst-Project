```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style='whitegrid')
```

## ðŸ“¥ Load and Clean the Sales Data

```python
# Load CSV file
df = pd.read_csv('salesdata-old.csv')

# Keep only relevant columns
df = df[['Row Labels', 'Sum of Sales']]
df.columns = ['Category', 'Sales']

# Preview
df.head()
```

## ðŸ“Š Sales by Category (Bar Chart)

```python
plt.figure(figsize=(10,6))
df_sorted = df.sort_values(by='Sales', ascending=True)
sns.barplot(x='Sales', y='Category', data=df_sorted, hue='Category', palette='Blues_r', legend=False)
plt.title('Sales by Category', fontsize=16)
plt.xlabel('Sales')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

## ðŸ¥§ Top 5 Categories by Sales (Pie Chart)

```python
top5 = df.sort_values(by='Sales', ascending=False).head(5)
plt.figure(figsize=(7,7))
plt.pie(top5['Sales'], labels=top5['Category'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 5 Categories by Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample monthly sales data
sample_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1200, 1500, 1800, 1700, 1600, 2000, 
              2200, 2100, 1900, 2300, 2500, 2400]
}

# Create DataFrame
monthly_df = pd.DataFrame(sample_data)

# Plot area chart
plt.figure(figsize=(12, 6))
plt.fill_between(monthly_df['Month'], monthly_df['Sales'], color='skyblue', alpha=0.5)
plt.plot(monthly_df['Month'], monthly_df['Sales'], marker='o', color='blue')
plt.title('ðŸ“ˆ Monthly Sales Trend (Sample Data)', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```
